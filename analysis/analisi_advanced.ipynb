# Altre analisi usando SQLite

```python
import pandas as pd
import sqlite3
```

## LETTURA LEGENDA

```python
df_naturaincidenti=pd.read_csv("../data/descriptions/natura_incidenti.csv", sep=";", encoding="iso8859_15", index_col=False)
```

## SQLITE APPROACH

```python
conn=sqlite3.connect(":memory:")
```

```python
conn.execute("DROP TABLE IF EXISTS naturaincidenti")
df_naturaincidenti.to_sql("naturaincidenti", conn, index=False)
```

```python
result=pd.read_sql_query("SELECT * FROM naturaincidenti", conn)
```

```python
result
```

```python
df_dati=pd.read_csv("../data/after_preprocessing/after_preprocessing.csv", sep=",", encoding="iso8859_1")
```

```python
conn.execute("DROP TABLE IF EXISTS dati")
df_dati.to_sql("dati", conn)
```

```python
result=pd.read_sql_query("SELECT * FROM dati", conn, index_col=None)
```

```python
result
```

```python
conn.execute("DROP TABLE IF EXISTS naturaincidenti_conteggio")
conn.execute("CREATE TABLE naturaincidenti_conteggio AS SELECT Id, Tipologia, COUNT(*) Conteggio FROM naturaincidenti JOIN dati ON naturaincidenti.Id=dati.natura_incidente GROUP BY dati.natura_incidente")
```

## Conteggio tipologia di incidente

```python
result_tipo=pd.read_sql_query("SELECT * FROM naturaincidenti_conteggio ORDER BY Conteggio DESC", conn)
```

```python
result_tipo
```

## Analisi del numero di lesi per tipologia di incidente

```python
result_lesi=pd.read_sql_query("SELECT naturaincidenti_conteggio.Tipologia, naturaincidenti_conteggio.Conteggio, SUM(morti_entro_24_ore+morti_entro_30_giorni+feriti) Lesi, 1.0*(SUM(morti_entro_24_ore+morti_entro_30_giorni+feriti))/naturaincidenti_conteggio.Conteggio Media FROM naturaincidenti_conteggio JOIN dati ON naturaincidenti_conteggio.Id=dati.natura_incidente GROUP BY dati.natura_incidente ORDER BY Conteggio DESC", conn)
```
