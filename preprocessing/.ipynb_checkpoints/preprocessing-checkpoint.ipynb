# Preprocessing di base

```python
import pandas as pd
import os
```

```python
df_full=pd.read_csv("../data/raw/INCSTRAD_Microdati_Anno_2016.txt", sep="\t", encoding="iso8859_15")
```

```python
df_full
```

```python
print("Numero di righe full dataset: ", len(df_full.index))
```

## ESTRAZIONE DEI DATI DI MILANO

```python
df_milano=df_full[(df_full['provincia']==15) & (df_full['comune']==146)]
```

```python
df_milano
```

```python
print("Numero di righe relative a Milano:", len(df_milano.index))
```

## PER IL MOMENTO LASCIAMO STARE TANGENZIALE E AUTOSTRADA - QUALITA' LOCALIZZAZIONE DA VERIFICARE E CI INTERESSANO MENO

```python
df_milano_simplified=df_milano[(~df_milano['descrizione_strada'].str.contains('TANGENZIALE')) & (~df_milano['descrizione_strada'].str.contains('AUTOSTRADA'))]
```

```python
print("Numero di righe relative a Milano - filtered version:" ,len(df_milano_simplified.index))
```

## SALVATAGGIO DEL DATASET FILTRATO

```python
df_milano_simplified.to_csv("../data/after_preprocessing/after_preprocessing.csv")
```
